## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:
- inspect relevant columns
- figure out any reasonable imputations
- filter the dataset

```python
# filter by age to get keep accidents from 1983 onwards
df["Event.Date"] = pd.to_datetime(df["Event.Date"], errors="coerce")
df["Event.Year"] = df["Event.Date"].dt.year
df = df[df["Event.Year"] >= 1983]
df
```

```python
# Keep only airplanes
df = df[df["Aircraft.Category"].str.contains("Airplane", case=False, na=False)]
df
```

```python
#Keeping only professional builds
df = df[df["Amateur.Built"].str.upper() == "NO"] 
df
```

Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
df.columns.tolist()
```

Adding the total number of injured and unijured passengers in order to get the total number of paasgers.
Dividing the sum of Fatl injuries and serious injuries by total number of passengers to get the serious/fatal rate.

```python
#Estimate total passengers
df["Total_Passengers"] = df["Total.Fatal.Injuries"] + df["Total.Minor.Injuries"] + df["Total.Serious.Injuries"] + df["Total.Uninjured"]

# Fraction of serious/fatal injuries
df["SeriousFatal_Rate"] = (df["Total.Fatal.Injuries"] + df["Total.Serious.Injuries"]) / df["Total_Passengers"]
df
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
df["Make"] = df["Make"].str.strip().str.upper()
make_counts = df["Make"].value_counts()
valid_makes = make_counts[make_counts >= 50].index
df = df[df["Make"].isin(valid_makes)]
df = df.dropna(subset=["Model"]).copy()
# Unique identifier
df["Plane_ID"] = df["Make"] + "_" + df["Model"].fillna("")
df
```

```python
#Inspecting the counts
print(df["Model"].value_counts().head(20))
model_counts = df.groupby("Make")["Model"].value_counts()
print(model_counts.head(30))
```

### Column Removal
- inspect the dataframe and drop any columns that have too many NaNs

```python
na_counts = df.isna().sum().sort_values(ascending=False)
na_counts
```

To correctly determine which columns to drop, we will find the %age of missing values relative to each column and drop columns that contain < 40% missing values.

```python
na_percent = (df.isna().sum() / len(df)) * 100
na_percent
```

```python
#Drop columns with more than 40% NaNs
threshold = 0.4
df = df.loc[:, df.isna().mean() < threshold]
print("Remaining columns:", df.columns.tolist())
print("Shape after dropping:", df.shape)
```
